```python
#import necessary libraries
from fastai import *
from fastai.vision import *
from torch.utils.data.sampler import BatchSampler, WeightedRandomSampler
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = datapath4file('resampled_random_db')
tfms = get_transforms(max_rotate=50)
#fnames = get_image_files(path)
#fnames
#AN, SA, SU, DI, NE, HA, FE
path.ls()
#data = ImageDataBunch.from_folder(path,ds_tfms=get_transforms(), valid_pct=0.2, size=224)
data = (ImageList.from_folder(path)
        .split_by_rand_pct(valid_pct=0.2)
        .label_from_folder()
        .transform(tfms, size=224)
        .databunch()).normalize(imagenet_stats)
```

```python
print('train set: ' + str(len(data.train_ds)))
print('valid set: ' + str(len(data.valid_ds)))
```

```python
learn = cnn_learner(data, models.resnet34, metrics=accuracy)
```

```python
learn.fit_one_cycle(4)
#learn.save('first_test_JAFFEETFEID')
# interp = ClassificationInterpretation.from_learner(learn)
# interp.plot_top_losses(18, figsize=(15,11))
```

```python
learn.lr_find()
learn.recorder.plot(suggestion=True)
```

```python
learn.unfreeze()
learn.fit_one_cycle(8, max_lr=slice(3.31E-06, 1e-02))
```

```python
learn.export()
learn.save('final77')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(5,5), dpi=60)
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
learn.export()
```
